```python
df_filter_1 = orders[(orders.o_comment.str.contains("^.*?special.*?requests.*?$",regex=True) == False)]
df_filter_1 = df_filter_1[['o_custkey', 'o_comment', 'o_orderkey']]
df_filter_2 = customer[['c_custkey']]
df_merge_1 = df_filter_1.merge(df_filter_2, left_on=['o_custkey'], right_on=['c_custkey'], how="right", sort=False)
df_group_1 = df_merge_1 \
    .groupby(['c_custkey'], sort=False, as_index=False) \
    .agg(
        c_count=("o_orderkey", "count"),
    )
df_group_1 = df_group_1[['c_count', 'c_custkey']]
df_group_2 = df_group_1 \
    .groupby(['c_count'], sort=False, as_index=False) \
    .agg(
        custdist=("c_count", "count"),
    )
df_group_2 = df_group_2[['custdist', 'c_count']]
df_sort_1 = df_group_2.sort_values(by=['custdist', 'c_count'], ascending=[False, False])
```

```python
df_scan_1 = customer[['c_custkey']]
df_scan_2 = orders[(orders.o_comment.str.contains('^.*?special.*?requests.*?$', regex=True) == False)]
df_scan_2 = df_scan_2[['o_orderkey', 'o_custkey', 'o_comment']]
df_join_1 = df_scan_1.merge(df_scan_2, left_on=['c_custkey'], right_on=['o_custkey'], how='outer', sort=True)
df_group_1 = df_join_1 \
    .groupby(['c_custkey'], sort=False, as_index=False) \
    .agg(
        o_orderkeycount=('o_orderkey', 'count'),
        count=('o_custkey', 'count'),
    )
df_group_2 = df_group_1 \
    .groupby(['o_orderkeycount'], sort=False, as_index=False) \
    .agg(
        count2=('c_custkey', 'count'),
    )
df_sort_1 = df_group_2.sort_values(
    by=['count2', 'o_orderkeycount'],
    ascending=[False, False]
)
df_output_1 = df_sort_1.rename(columns={'o_orderkeycount': 'c_count', 'count2': 'custdist'})
df_output_1 = df_output_1[['c_count', 'custdist']]
```
