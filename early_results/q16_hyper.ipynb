```python
import pandas as pd
def q16(partsupp, part, supplier):
    df_scan_1 = supplier[supplier.s_comment.str.contains('^.*?Customer.*?Complaints.*?$', regex=True)]
    df_scan_1 = df_scan_1[['s_suppkey', 's_comment']]
    df_scan_2 = part[(part.p_brand != 'Brand#45') & (part.p_type.str.contains('^MEDIUM POLISHED.*?$', regex=True) == False) & part.p_size.isin([3, 9, 14, 19, 23, 36, 45, 49])]
    df_scan_2 = df_scan_2[['p_partkey', 'p_brand', 'p_type', 'p_size']]
    df_scan_3 = partsupp[['ps_partkey', 'ps_suppkey']]
    df_join_1 = df_scan_2.merge(df_scan_3, left_on=['p_partkey'], right_on=['ps_partkey'], how='inner', sort=False)
    df_join_2 = df_scan_1.merge(df_join_1, left_on=['s_suppkey'], right_on=['ps_suppkey'], how='outer', sort=False, indicator=True)
    df_join_2 = df_join_2[df_join_2._merge == 'right_only'].drop('_merge', axis = 1)
    df_group_1 = df_join_2 \
        .groupby(['p_brand', 'p_type', 'p_size'], sort=False, as_index=False) \
        .agg(
            ps_suppkeycount=('ps_suppkey', lambda x : x.nunique()),
        )
    df_output_1 = df_group_1.rename(columns={'ps_suppkeycount': 'supplier_cnt'})
    df_output_1 = df_output_1[['p_brand', 'p_type', 'p_size', 'supplier_cnt']]
    return df_output_1
```
