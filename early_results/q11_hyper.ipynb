```python
df_scan_1 = nation[(nation.n_name == 'GERMANY')]
df_scan_1 = df_scan_1[['n_nationkey', 'n_name']]
df_scan_2 = supplier[['s_suppkey', 's_nationkey']]
df_join_1 = df_scan_1.merge(df_scan_2, left_on=['n_nationkey'], right_on=['s_nationkey'], how='inner', sort=False)
df_scan_3 = partsupp[['ps_suppkey', 'ps_availqty', 'ps_supplycost']]
df_join_2 = df_join_1.merge(df_scan_3, left_on=['s_suppkey'], right_on=['ps_suppkey'], how='inner', sort=False)
df_join_2['ps_supplycostps_availqty'] = df_join_2.ps_supplycost * df_join_2.ps_availqty
df_aggr_1 = pd.DataFrame()
df_aggr_1['ps_supplycostps_availqtysum'] = [df_join_2.ps_supplycostps_availqty.sum()]
df_aggr_1['ps_supplycostps_availqtysum1'] = df_aggr_1.ps_supplycostps_availqtysum * 0.0001
df_scan_4 = nation[(nation.n_name == 'GERMANY')]
df_scan_4 = df_scan_4[['n_nationkey', 'n_name']]
df_scan_5 = supplier[['s_suppkey', 's_nationkey']]
df_join_3 = df_scan_4.merge(df_scan_5, left_on=['n_nationkey'], right_on=['s_nationkey'], how='inner', sort=False)
df_scan_6 = partsupp[['ps_partkey', 'ps_suppkey', 'ps_availqty', 'ps_supplycost']]
df_join_4 = df_join_3.merge(df_scan_6, left_on=['s_suppkey'], right_on=['ps_suppkey'], how='inner', sort=False)
df_join_4['ps_supplycostps_availqty0'] = df_join_4.ps_supplycost * df_join_4.ps_availqty
df_group_2 = df_join_4 \
    .groupby(['ps_partkey'], sort=False, as_index=False) \
    .agg(
        ps_supplycostps_availqtysum5=('ps_supplycostps_availqty0', 'sum'),
    )
    
print(df_group_2.columns)
print(df_aggr_1.columns)
df_join_5 = df_aggr_1.merge(df_group_2, how='cross', sort=False)
df_join_5 = df_join_5[(df_join_5.ps_supplycostps_availqtysum5 > df_join_5.ps_supplycostps_availqtysum1)]
df_sort_1 = df_join_5.sort_values(
    by=['ps_supplycostps_availqtysum5'],
    ascending=[False]
)
df_output_1 = df_sort_1.rename(columns={'ps_supplycostps_availqtysum5': 'value'})
df_output_1 = df_output_1[['ps_partkey', 'value']]
```
