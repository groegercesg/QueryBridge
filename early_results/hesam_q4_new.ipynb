```python
df_filter_1 = orders[(orders.o_orderdate >= '1993-07-01 00:00:00') & (orders.o_orderdate < '1993-10-01 00:00:00')]
df_filter_1 = df_filter_1[['o_orderkey', 'o_orderdate', 'o_orderpriority']]
df_sort_1 = df_filter_1.sort_values(by=['o_orderkey'], ascending=[True])
df_sort_1 = df_sort_1[['o_orderkey', 'o_orderdate', 'o_orderpriority']]
df_filter_2 = lineitem[lineitem.l_commitdate < lineitem.l_receiptdate]
df_filter_2 = df_filter_2[['l_orderkey']]
# Remove empty list
#df_filter_2 = df_filter_2.rename_axis().reset_index()
df_rename_1 = pd.DataFrame()
df_rename_1['l_orderkey'] = df_filter_2['l_orderkey']
df_group_1 = df_rename_1 \
    .groupby(['l_orderkey'], sort=False) \
    .last()
df_group_1.index = df_group_1.index.rename('l_orderkey')
df_group_1 = df_group_1.reset_index(level=0)
df_merge_1 = df_sort_1.merge(df_group_1, left_on=['o_orderkey'], right_on=['l_orderkey'], how="inner", sort=True)
df_merge_1 = df_merge_1[['o_orderkey', 'o_orderdate', 'o_orderpriority']]
df_group_2 = df_merge_1 \
    .groupby(['o_orderpriority'], sort=False) \
    .agg(
        count_o_orderdate=("o_orderdate", "count"),
    )
df_group_2['counto_orderdate'] = df_group_2.count_o_orderdate
df_group_2 = df_group_2[['counto_orderdate']]
df_sort_2 = df_group_2.sort_values(by=['o_orderpriority'], ascending=[True])
df_sort_2 = df_sort_2[['counto_orderdate']]
#return df_sort_2
```
